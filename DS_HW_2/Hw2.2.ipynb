```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("2017_jun_final.csv")
print(data)
```

```python
print(data.head())
```

```python
data.shape
```

```python
print(data.dtypes)
```

```python
missed = data.isnull().sum(axis=0) / len(data)
print(missed)
```

```python
columns = data.columns[data.isna().any()]

columns_drop = [col for col in columns if col != 'Язык.программирования']
data = data.drop(columns=columns_drop)
data
```

```python
data.replace(1, np.nan, inplace=True)
data.dropna(inplace=True)
data
```

```python
data.shape
```

```python
python_data = data[data["Язык.программирования"] == "Python"]
python_data
```

```python
python_data.shape
```

```python
position = python_data.groupby("Должность")
```

```python
salary = position["Зарплата.в.месяц"].agg(['min', 'max'])
salary
```

```python
def fill_avg_salary(row):
    return (row.min() + row.max()) / 2
salary["avg"] = salary.apply(fill_avg_salary, axis=1)
salary.sort_values('avg', ascending=False)
```

```python
salary['avg'].describe()
```

```python
salary.to_csv("salary.csv", index=False)
```

```python
sns.scatterplot(x='Общий.опыт.работы', y='Изменение.зарплаты.за.12.месяцев', data=data)
plt.xlabel('Общий опыт работы (лет)')
plt.ylabel('Изменение зарплаты за 12 месяцев')
plt.title('Зависимость изменения зарплаты от опыта работы')
plt.show()
```

```python
sns.histplot(data=data, x='Зарплата.в.месяц', kde=True)
plt.xlabel('Зарплата в месяц')
plt.ylabel('Частота')
plt.title('Распределение зарплат в месяц')
plt.show()
```

```python
sns.boxplot(x='Город', y='Зарплата.в.месяц', data=data)
plt.xlabel('Город')
plt.ylabel('Зарплата в месяц')
plt.title('Распределение зарплат по городам')
plt.xticks(rotation=90)
plt.show()
```
