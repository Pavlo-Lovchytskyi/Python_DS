```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import pandas as pd
import matplotlib.pyplot as plt
import os
os.environ["LOKY_MAX_CPU_COUNT"] = "4"
```

```python
data_2d = pd.read_csv('data_2d.csv').values
data_2d
```

```python
inertia_2d = []
for n_clusters in range(1, 6):
    kmeans = KMeans(n_clusters=n_clusters, n_init='auto')
    kmeans.fit(data_2d)
    inertia_2d.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), inertia_2d, marker='o')
plt.title('Silhouette Score для визначення оптимальної кількості кластерів')
plt.xlabel('Кількість кластерів')
plt.ylabel('Silhouette Score')
```

```python
kmeans_2d = KMeans(n_clusters=2, n_init='auto')  
kmeans_2d.fit(data_2d)
labels = kmeans_2d.labels_
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, cmap='viridis', edgecolors='r', s=50)
plt.title('2D Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
mnist = pd.read_csv('mnist.csv')
mnist_data = pd.DataFrame(mnist/255.0).values
mnist_data
```

```python
inertia_mnist = []
for n_clusters in range(1, 20):
    kmeans = KMeans(n_clusters=n_clusters, verbose=0)
    kmeans.fit(mnist_data)
    inertia_mnist.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.plot(range(1, 20), inertia_mnist, marker='o')
plt.title('Elbow Method for MNIST Dataset')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.show()
```

```python
pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(mnist_data)
kmeans_mnist = KMeans(n_clusters=10, n_init='auto')
kmeans_mnist.fit(mnist_pca)
labels_mnist = kmeans_mnist.labels_
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=labels_mnist, cmap='viridis', edgecolors='c', s=50)
plt.title('MNIST Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()
```
